```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
        return True
    
print(is_prime(29))
```

```python
def remove_duplicates(lst):
    unique = []
    for item in lst:
        if item not  in unique:
            unique.append(item)
            return unique
        
print(remove_duplicates([1,2,2,3,4,4,5]))
```

```python
def char_frequency(text):
    freq = {}
    for char in text:
        if char != " ":
            freq[char] = freq.get(char, 0) + 1
            return freq
        
print(char_frequency("python programming"))
```

```python
def dummy_mileage_kmpl(engine_l, weight_kg, tyre_in):
    mileage = 28 - 4 * engine_l - 0.006 * (weight_kg - 1000) - 0.7 * (tyre_in - 15)
    return max(5.0, mileage)


def city_highway_kmpl(engine_l, weight_kg, tyre_in):
    base = dummy_mileage_kmpl(engine_l, weight_kg, tyre_in)
    city = base * 0.85
    highway = base * 1.10
    return base, city, highway


# Fixed values (no input)
engine = 1.5
weight = 1200
tyre = 16

base, city, highway = city_highway_kmpl(engine, weight, tyre)

print("===== Mileage Report =====")
print("Engine:", engine, "L")
print("Weight:", weight, "kg")
print("Tyre:", tyre, "inch")
print("Base Mileage:", round(base, 2), "km/L")
print("City Mileage:", round(city, 2), "km/L")
print("Highway Mileage:", round(highway, 2), "km/L")
```

```python
num = int(input("Enter number: "))
power = len(str(num))
total = 0
temp = num

while temp > 0:
    digit = temp % 10
    total += digit ** power
    temp //= 10

if total == num:
    print("Armstrong Number")
else:
    print("Not Armstrong Number")
```

```python
numbers = [1, 2, 4, 5, 1, 7, 8]

seen = set()
duplicates = set()

for num in numbers:
    if num in seen:
        duplicates.add(num)
    else:
        seen.add(num)

print("Duplicate elements: ", list(duplicates))
```
