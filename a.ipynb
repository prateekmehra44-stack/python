```python
def top_k(items, k):
    return sorted(items, key=lambda x:x[1], reverse=True)[:k]
items = [("A", 90), ("B", 95), ("C", 95), ("D", 85)]
k = 2

result = top_k(items, k)
print(result)
```

```python
def count_by_category(pairs):
    d = {}
    for cat, words in pairs:
        d.setdefault(cat, {})
        d[cat].setdefault(words, 0)
        d[cat][words] += 1
        return d
```

```python
def missing_number(nums):
    n = len(nums)
    total = n * (n + 1) // 2
    return total - sum(nums)
```

```python
def rle_decode(s):
    result = ""
    i = 0
    n = len(s)

    while i < n:
        if not s[i].isdigit():
            raise ValueError("Invalid format")
        
        count_str = ""
        while i < n and s[i].isdigit():
            count_str += s[i]
            i += 1
            if i >= n:
                raise ValueError("Invalid format")
            
            count = int(count_str)

            if s[i] == "\\":
                i += 1
                if i >= n:
                        
                        raise ValueError("Invalid escape")
                    char = s[i]
                else:
                    char = s[i]
                    result += char * count
                    i += 1
                    return result
```
